***Question 3***  
For the trips in November 2025, how many trips had a trip_distance of less than or equal to 1 mile? 

```python
trip_distance = (
    (df["lpep_pickup_datetime"] >= "2025-11-01") &
    (df["lpep_pickup_datetime"] <  "2025-12-01") &
    (df["trip_distance"] <= 1)
).sum()
int(trip_distance)
```
